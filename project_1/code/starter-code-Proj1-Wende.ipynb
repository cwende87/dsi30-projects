Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. 2. <details><summary>Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Define function to calculate standard deviation. 
def stddev(values_list):
    sum = 0
    
    #obtain mean of values_list.
    mean = find_mean(values_list)
    
    #finding sum of the differences of all values from mean and square individual results. 
    for num in values_list:
        sum += (num-mean)**2
    
    stddev = (sum/len(values_list))**0.5
    
    return stddev


round(stddev(values), 2)
```

For comparing SAT and ACT scores, the College Board organisation provides conversion charts to show how ACT composite score compare with SAT total score. Additional variables called 'ACT equivalent' is added into the merged dataframe. 

```python
# reading concordance table
df = pd.read_csv('../data/concordance_table.txt')
df.head()
```

```python
# Preparing values to be initialise in dataframe.
concordance_table_values = [df.iloc[index][0].replace('–', ' ').split(' ') for index in df.index]
concordance_table_values
```

```python
# Creating a dataframe for concordance table
concordance_table = pd.DataFrame(concordance_table_values, columns = ['act', 'sat', 'sat_lower', 'sat_upper'])
concordance_table.info()
```

```python
# Converting values to int or float
concordance_table['act'] = concordance_table['act'].map(float)

for column in concordance_table.columns[1:]:
    concordance_table[column] = concordance_table[column].map(int)

concordance_table.info()
```

```python
# Define function to translate SAT total to ACT equivalent composite score

def to_act(series):
    
    out = []
    
    for value in series:
        for i in concordance_table.index:
            upper = concordance_table['sat_upper'][i]
            lower = concordance_table['sat_lower'][i]
            if value>=lower and value<=upper:
                    out.append(int(concordance_table_values[i][0]))
    
    return out
```

```python
# Create new variables in the std_test_merged dataframe to be used for comparison. 
std_test_merged['sat_2017_act_equivalent'] = to_act(std_test_merged['sat_2017_total'])
std_test_merged['sat_2018_act_equivalent'] = to_act(std_test_merged['sat_2018_total'])
std_test_merged['sat_2019_act_equivalent'] = to_act(std_test_merged['sat_2019_total'])
```

    - Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

```python
# Define subplot histogram function. It takes the names of 2+ columns and subplot histograms.

def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, ylabels, plot_title, bin):
    nrows = int(np.ceil(len(list_of_columns)/3)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, sharex=False , sharey=True, figsize = (15,10))
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    plt.suptitle(plot_title, fontsize=20, y=1.05)
    plt.xticks()
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        values, bins, bars = ax[i].hist(dataframe[column])
        ax[i].set_title(list_of_titles[i], fontsize=16, y=1.01)   # feel free to add more settings
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=12)  # Set titles, labels, etc here for each subplot
        ax[i].set_ylabel(ylabels, fontsize=12)
        ax[i].axvline(dataframe[column].median(), color='r', linestyle='dashed', linewidth=1, label='median')
        ax[i].axvline(dataframe[column].mean(), color='k', linestyle='dashed', linewidth=1, label = 'mean')
        ax[i].legend()
        ax[i].bar_label(bars, fontsize=10) 
        
    plt.tight_layout()   
```

```python
# Define subplot scatter plot function.

def subplot_scatter(dataframe, variable1, variable2, list_of_subplottitles, xlabels, ylabels, plot_title):
    nrows = int(np.ceil(len(variable1)/3)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, sharex=False , sharey=True, figsize = (10,5))
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    plt.suptitle(plot_title, fontsize=20, y=1.05)
    plt.xticks()
            
    for i, column in enumerate(variable1): # Gives us an index value to get into all our lists
        ax[i].scatter(dataframe[variable1[i]], dataframe[variable2[i]])
        ax[i].set_title(list_of_subplottitles[i], fontsize=16, y=1.01)   # feel free to add more settings
        ax[i].set_xlabel(xlabels, fontsize=12)  # Set titles, labels, etc here for each subplot
        ax[i].set_ylabel(ylabels, fontsize=12)
                
        print(f"Figure {i+1} correlation is {round(dataframe[variable1[i]].corr(dataframe[variable2[i]]),2)}")

    plt.tight_layout() 
```
